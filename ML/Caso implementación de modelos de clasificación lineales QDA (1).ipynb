### 4. Estimacin modelo NBC (20 puntos)
Predigan usando Naive Bayes Classifier y evalen su poder predictivo en la muestra de test. Concluyan sobre el poder predictivo del modelo estimado.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc

yTest_binary = np.where(yTest == 'Up', 1, 0)
nbc_model = GaussianNB()
nbc_model.fit(XTrain, yTrain)
yPred_nbc = nbc_model.predict(XTest)
y_prob_nbc = nbc_model.predict_proba(XTest)[:, 1]
```

```python
accuracy_nbc = accuracy_score(yTest, yPred_nbc)
conf_matrix_nbc = confusion_matrix(yTest, yPred_nbc)
classification_rep_nbc = classification_report(yTest, yPred_nbc)

fpr, tpr, _ = roc_curve(yTest_binary, y_prob_nbc)
roc_auc = auc(fpr, tpr)

```

```python
plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.show()

print(f'Accuracy del modelo Nave Bayes: {accuracy_nbc:.4f}')
print('\nMatriz de Confusin:')
print(conf_matrix_nbc)
print('\nReporte de Clasificacin:')
print(classification_rep_nbc)
```

# Anlisis y Conclusin
El modelo Nave Bayes obtuvo una precisin del 52.78%, lo que indica un rendimiento apenas superior al azar. La matriz de confusin revela que el modelo tiene dificultades para clasificar correctamente la clase 'Down', con un recall de solo el 29%. En cambio, para la clase 'Up', el modelo muestra un mejor desempeo con un recall del 77%, aunque la precisin es baja. Esto sugiere que el modelo tiende a clasificar ms observaciones como 'Up' en comparacin con 'Down'.

La curva ROC y el valor del AUC proporcionan una evaluacin adicional del rendimiento del modelo. Si el AUC es cercano a 0.5, el modelo no discrimina bien entre clases.
